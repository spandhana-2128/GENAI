```python
from transformers import pipeline
import matplotlib.pyplot as plt
```

```python
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
diary_entries = {
    "Monday": "I felt motivated and productive today.",
    "Tuesday": "The workload was stressful and tiring.",
    "Wednesday": "It was an average day.",
    "Thursday": "I am happy with my project progress.",
    "Friday": "I felt exhausted and demotivated.",
    "Saturday": "Relaxed and enjoyed time with friends.",
    "Sunday": "Feeling calm and prepared for the week."
}
```

```python
mood_scores = {}

for day, text in diary_entries.items():
    result = sentiment_analyzer(text)[0]
    mood_scores[day] = 1 if result["label"] == "POSITIVE" else -1
```

```python
days = list(mood_scores.keys())
scores = list(mood_scores.values())

plt.plot(days, scores, marker='o')
plt.xlabel("Day")
plt.ylabel("Mood Score")
plt.title("Weekly Mood Tracker")
plt.show()
```
